```python
import pandas
from markovclick.models import MarkovClickstream
from markovclick.viz import visualise_markov_chain
import matplotlib.pyplot as pyplot
import seaborn
```

```python
dataframe=pandas.read_csv('finalMM_data.csv')
```

```python
dataframe.set_index(dataframe.columns[0],inplace=True)
```

```python
data=[]
for i in range(75):
    x=dataframe.values[i]
    x=x[~pandas.isnull(x)]
    data.append(x)
data
```

```python
m=MarkovClickstream(data)
fig=pyplot.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
s=seaborn.heatmap(m.prob_matrix, xticklabels=m.pages, yticklabels=m.pages,cmap="YlGnBu",annot=True)
pyplot.show()
```

```python
graph = visualise_markov_chain(m)
graph
```
